```python
import seaborn as sns
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
iris=px.data.iris()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip')
```

```python
# stripplot
# axes level
```

```python
sns.stripplot(data=tips,x='day',y='total_bill',jitter=False)
```

```python
sns.catplot(kind='strip',data=tips,x='day',y='total_bill')
```

```python
sns.catplot(kind='strip',data=tips,x='day',y='total_bill',jitter=.2,hue='sex')
```

```python
sns.boxplot(data=tips,x='day',y='total_bill')
```

```python
sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker')
```

```python
import numpy as np
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
g=sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill',)
g.add_legend()
```

```python
sns.pairplot(iris,hue='species')
```

```python
#pairgrid
g=sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
```

```python
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatte)
```

```python
sns.jointplot(data=tips,kind='hist',x='total_bill',y='tip',hue='sex')
```

```python
g=sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.histplot)
```

```python
g=sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.kdeplot,sns.jointplot)
```
